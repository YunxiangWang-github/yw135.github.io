## Imports

```python
import os
import pandas as pd
import altair as alt
```

## Read in Data

```python
alt.data_transformers.disable_max_rows()

DATA_URL = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
OUT_DIR = "../assets/hw5"
os.makedirs(OUT_DIR, exist_ok=True)

df = pd.read_csv(DATA_URL)

df = df[['Agency Name', 'County', 'Square Footage', 'Bldg Status']].dropna()
df.columns = ['agency', 'county', 'sqft', 'status']

df.head()

```

## Visualization 1:

```python
import numpy as np
import altair as alt

top10 = df['county'].value_counts().nlargest(10).index
df1 = df.copy()
df1['bucket'] = np.where(df1['county'].isin(top10), 'Top 10', 'Other')

legend_sel = alt.selection_point(fields=['agency'], bind='legend')

view = alt.param(
    name='view',
    bind=alt.binding_select(options=['Top 10', 'All', 'Other'], name='View: '),
    value='Top 10'
)

chart1 = (
    alt.Chart(df1)
    .mark_bar()
    .encode(
        x=alt.X('county:N', sort='-y', title='County',
                axis=alt.Axis(labelAngle=-60)),
        y=alt.Y('count():Q', title='Building Count'),
        color=alt.Color('agency:N', title='Agency'),
        opacity=alt.condition(legend_sel, alt.value(0.95), alt.value(0.2)),
        tooltip=['county','agency','count()']
    )
    .add_params(legend_sel, view)
    .transform_filter("(view == 'All') || (datum.bucket == view)")
    .properties(width=700, height=400,
                title='Building Counts by County and Agency')
)

chart1.save(f"{OUT_DIR}/chart1.json")
"Saved chart1.json"
```

## Visualization 2:

```python
counties = sorted(df['county'].dropna().unique().tolist())

dropdown = alt.param(
    name='county_sel',
    bind=alt.binding_select(options=counties, name='Select County: '),
    value=counties[0]
)

chart2 = (
    alt.Chart(df)
      .mark_bar()
      .encode(
          x=alt.X('agency:N', sort='-y', title='Agency'),
          y=alt.Y('mean(sqft):Q', title='Average Square Footage'),
          color=alt.Color('agency:N', legend=None),
          tooltip=['agency', 'mean(sqft):Q']
      )
      .add_params(dropdown)
      .transform_filter("datum.county == county_sel")
      .properties(width=700, height=400, title='Average Building Size by Agency (Filtered by County)')
)

chart2.save(f"{OUT_DIR}/chart2.json")
f"Saved {OUT_DIR}/chart2.json"
```
